```python
import pickle
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt

from pathlib import Path
```

```python
INSTANCE_PATH = Path("data")
RESULTS_PATH = Path("results")
```

```python
with open(RESULTS_PATH / "vrp_4_25.pkl", 'rb') as f:
    results = pickle.load(f)
    
with open(INSTANCE_PATH / 'instances.pkl', 'rb') as f:
    instances = pickle.load(f)
```

```python
dataset = [(x, y['routes'], y['classifications']) for (x, y) in zip(instances, results)]
```

```python
locations, routes, _ = dataset[0]

plt.scatter(locations[:, 0], locations[:, 1])
plt.scatter(locations[0, 0], locations[0, 1]) # depot

# plot the route
for i, path in routes.items():
    path = [0] + path + [0]
    path_locations = np.array([locations[loc, :2] for loc in path])
    plt.plot(path_locations[:, 0], path_locations[:, 1])
```
